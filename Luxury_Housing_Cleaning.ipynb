```python
## make a copy of the dataset before cleaning
df_copy = df.copy()

df_copy.shape
```

```python
df_copy.duplicated().sum()

df_copy = df_copy.drop_duplicates(ignore_index=True)

df_copy.duplicated().sum()

df_copy.shape
```

```python
## Configuration — convert to uppercase
df_copy['Configuration'] = df_copy['Configuration'].str.strip().str.upper()

df_copy['Configuration'].value_counts()
```

```python
## Micro_Market — convert to title case
df_copy['Micro_Market'] = df_copy['Micro_Market'].str.strip().str.title()

df_copy['Micro_Market'].value_counts()
```

```python
## NRI_Buyer — standardize yes/no
df_copy['NRI_Buyer'] = df_copy['NRI_Buyer'].str.strip().str.title()

df_copy['NRI_Buyer'].value_counts()
```

```python
## strip whitespace from other text columns
df_copy['Developer_Name'] = df_copy['Developer_Name'].str.strip()
df_copy['Buyer_Type'] = df_copy['Buyer_Type'].str.strip()
df_copy['Transaction_Type'] = df_copy['Transaction_Type'].str.strip()
df_copy['Possession_Status'] = df_copy['Possession_Status'].str.strip()
df_copy['Sales_Channel'] = df_copy['Sales_Channel'].str.strip()
df_copy['Buyer_Comments'] = df_copy['Buyer_Comments'].str.strip()
```

```python
# need to remove ₹ and Cr, then convert to float

df_copy['Ticket_Price_Cr'] = df_copy['Ticket_Price_Cr'].astype(str)
df_copy['Ticket_Price_Cr'] = df_copy['Ticket_Price_Cr'].str.replace('₹', '')
df_copy['Ticket_Price_Cr'] = df_copy['Ticket_Price_Cr'].str.replace('Cr', '')
df_copy['Ticket_Price_Cr'] = df_copy['Ticket_Price_Cr'].str.strip()
df_copy['Ticket_Price_Cr'] = df_copy['Ticket_Price_Cr'].replace('nan', np.nan)

## convert to numeric
df_copy['Ticket_Price_Cr'] = pd.to_numeric(df_copy['Ticket_Price_Cr'], errors='coerce')

df_copy['Ticket_Price_Cr'].dtype

print("Nulls after conversion:", df_copy['Ticket_Price_Cr'].isnull().sum())
print("Negative values:", (df_copy['Ticket_Price_Cr'] < 0).sum())
```

```python
## Unit_Size_Sqft: replace -1 with NaN
df_copy.loc[df_copy['Unit_Size_Sqft'] <= 0, 'Unit_Size_Sqft'] = np.nan


## Ticket_Price_Cr: replace negative prices with NaN
df_copy.loc[df_copy['Ticket_Price_Cr'] < 0, 'Ticket_Price_Cr'] = np.nan
```

```python
df_copy.isnull().sum()
```

```python
# Unit_Size_Sqft
df_copy['Unit_Size_Sqft'].median()

df_copy['Unit_Size_Sqft'] = df_copy['Unit_Size_Sqft'].fillna(df_copy['Unit_Size_Sqft'].median())

```

```python
# Ticket_Price_Cr
df_copy['Ticket_Price_Cr'].median()

df_copy['Ticket_Price_Cr'] = df_copy['Ticket_Price_Cr'].fillna(df_copy['Ticket_Price_Cr'].median())

```

```python
# Amenity_Score
df_copy['Amenity_Score'].median()

df_copy['Amenity_Score'] = df_copy['Amenity_Score'].fillna(df_copy['Amenity_Score'].median())

```

```python
## Text column — fill with placeholder
df_copy['Buyer_Comments'] = df_copy['Buyer_Comments'].fillna('No Comment')

## verify all nulls are handled
df_copy.isnull().sum()
```

```python
df_copy.info()

df_copy.describe()
```

```python
# Feature engg
# Price_Per_Sqft = (Price in rupees) / (Area in sqft)

df_copy['Price_Per_Sqft'] = (df_copy['Ticket_Price_Cr'] * 10000000) / df_copy['Unit_Size_Sqft']
df_copy['Price_Per_Sqft'] = df_copy['Price_Per_Sqft'].round(2)

df_copy['Price_Per_Sqft'].describe()
```

```python
df_copy['Purchase_Quarter'].unique()

df_copy['Purchase_Date'] = pd.to_datetime(df_copy['Purchase_Quarter'], errors='coerce')

df_copy['Purchase_Year'] = df_copy['Purchase_Date'].dt.year
df_copy['Purchase_Qtr_Num'] = df_copy['Purchase_Date'].dt.quarter
df_copy['Quarter_Label'] = 'Q' + df_copy['Purchase_Qtr_Num'].astype(str) + ' ' + df_copy['Purchase_Year'].astype(str)

# ----- Drop intermediate column -----

df_copy = df_copy.drop(columns=['Purchase_Date'])
```

```python
df_copy['Price_Tier'] = pd.cut(
    df_copy['Ticket_Price_Cr'],
    bins=[0, 5, 10, 15, 20, df_copy['Ticket_Price_Cr'].max() + 1],
    labels=['Entry Luxury', 'Mid Luxury', 'Premium', 'Ultra Premium', 'Super Luxury']
)
```

```python
# np.where(condition, value_if_true, value_if_false)
# useful for calculating percentages later

df_copy['NRI_Flag'] = np.where(df_copy['NRI_Buyer'] == 'Yes', 1, 0)

df_copy['NRI_Flag'].value_counts()
```
